# Web Scraping seach_tambon

# search_tambon_all_province

```python
from urllib.parse import quote
from gazpacho import get, Soup
import pandas as pd

# Fetch the main page HTML for 'search-tambon'
main_url = "https://healthgate.hss.moph.go.th/search-tambon"
main_html = get(main_url)
main_soup = Soup(main_html)

# Extract province codes from the <select> element
province_codes = [option.attrs['value'] for option in main_soup.find('select', {'name': 'province'}).find('option', mode='all')[1:]]

all_data = []

for code in province_codes:
    # Build the URL for each province code
    url = f"https://healthgate.hss.moph.go.th/search/tambon?province={code}"
    
    html = get(url)
    soup = Soup(html)

    # Extract the data for each tambon
    table_body = soup.find('tbody')
    rows = table_body.find('tr', mode='all') if table_body else []
    for row in rows:
        cells = row.find('td', mode='all')
        row_data = [cell.text for cell in cells]
        all_data.append(row_data)

# Define columns for the DataFrame based on the expected output
columns = ["จังหวัด", "อำเภอ", "ตำบล", "Login Code"]

# Convert all collected data into a DataFrame
df = pd.DataFrame(all_data, columns=columns)
```
