# Web Scraping seach_school

# search_school_all_province

```python
from urllib.parse import quote
from gazpacho import get, Soup
import pandas as pd

# Fetch the main page HTML
main_url = "https://healthgate.hss.moph.go.th/search-school"
main_html = get(main_url)
main_soup = Soup(main_html)

# Extract province names from the <select> element
province_names = [option.text for option in main_soup.find('select', {'id': 'search_province'}).find('option', mode='all')[1:]]

all_data = []

for name in province_names:
    # Use the province name for URL construction
    url = f"https://healthgate.hss.moph.go.th/search/school?search_province={quote(name)}"
    
    html = get(url)
    soup = Soup(html)

    # Extract school data
    table_body = soup.find('tbody')
    rows = table_body.find('tr', mode='all') if table_body else []
    for row in rows:
        cells = row.find('td', mode='all')
        row_data = [cell.text for cell in cells]
        all_data.append(row_data)

columns = ["จังหวัด", "อำเภอ", "ตำบล", "ชื่อโรงเรียน", "Login Code", "รหัส รพ.สต"]

# Convert all collected data into a DataFrame
df = pd.DataFrame(all_data, columns=columns)
```
